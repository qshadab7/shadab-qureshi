# Importing cust_account csv file

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('cust_acc')
print(data.shape)
data.head()
```

```python
data.dtypes
```

```python
data=data.loc[:,['customer_no','acct_type','cur_balance_amt','amt_past_due','creditlimit','cashlimit']]
```

```python
data.head()
```

```python
data.duplicated().sum()
```

```python
data.fillna(0,inplace=True)
```

```python
data.isnull().sum()
```

```python
df=data.groupby(['customer_no'])
df.head()
```

```python
acc_cust=pd.DataFrame(df.sum()).reset_index()
print(acc_cust.shape)
acc_cust.head()
```

```python
acc_cust.to_csv('final_accu',index=False)
```

# Importing cust_enquiry csv file

```python
data=pd.read_csv('cust_enq')
print(data.shape)
data.head()
```

```python
df=data.loc[:,['customer_no','enq_amt']]
```

```python
df.head(2)
```

```python
df0=df.groupby(['customer_no'])
df0.head()
```

```python
df1=pd.DataFrame(df0.sum()).reset_index()
df1.head()
```

```python
df2=pd.DataFrame(df0.count()).reset_index()
df2.head(2)
```

```python
cust=pd.merge(df2,df1,on='customer_no',how='inner')
cust.head()
```

```python
cust.duplicated().sum()
```

```python
cust.isnull().sum()
```

```python
cust.to_csv('final_enqu',index=False)
```
