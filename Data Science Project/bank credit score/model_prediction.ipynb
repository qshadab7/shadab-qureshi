```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data=pd.read_csv('final_data')
print(data.shape)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder()
```

```python
df=['feature_1','feature_5','feature_11','feature_23','feature_27','feature_32','feature_33',
   'feature_36','feature_37','feature_46','feature_48','feature_58','feature_59','feature_60',
   'feature_62','feature_72','feature_79']
for i in df:
    data[[i]]=la.fit_transform(data[[i]])
```

```python
data.head()
```

```python
from collections import Counter
Counter(data.Bad_label)
```

```python
x=data.iloc[:,:-1]
y=data.Bad_label
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=52,
                            max_depth=5,
                            max_features=9,
                            random_state=12)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
model=XGBClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
